```python
url = "https://www2.census.gov/geo/docs/maps-data/data/gazetteer/2019_Gazetteer/2019_gaz_tracts_39.txt"
latitude = []
longitude = []
Fips = []
file = urllib. request. urlopen(url)
for line in file:
    decoded_line = line. decode("utf-8")
    list_of_values = decoded_line.split()
    latitude.append(list_of_values[6])
    longitude.append(list_of_values[7])
    Fips.append(list_of_values[1])
final_latitude = latitude[1:]
final_longitude = longitude[1:]
final_Fips = Fips[1:]
```

### Since the first five numbers in a FIPS code refer to the state and county, I filtered out all rows which did not start with the numbers 39153, meaning they were not in Summit County, Ohio.

### Since the last six digits of the FIPS code is the census tract number times 100, I removed the first five digits from each FIPS code and divided it by 100 so it would match the census tract numbers I already had.

```python
def shorten_fips(row):
    return(row[5:])
census_tracts['fips number']=census_tracts['fips number'].apply(shorten_fips)
census_tracts
```

```python
def divide_by_100(row):
    row = float(row)
    return(row/100)
census_tracts['fips number']=census_tracts['fips number'].apply(divide_by_100)
census_tracts
```

### I then used the latitude and longitude of all libraries and the center of each census tract to determine the distance to the closest library for each census tract.

```python
census_tract_lat = list(census_tracts_in_akron['latitude'])
census_tract_long = list(census_tracts_in_akron['longitude'])
library_lat = list(library_info['latitude'])
library_long = list(library_info['longitude'])
```

```python
# uses formula described at https://www.geeksforgeeks.org/program-distance-two-points-earth/
from math import radians, cos, sin, asin, sqrt
def determine_distance(lat1,lat2,long1,long2):
    lat1 = radians(float(lat1))
    lat2 = radians(float(lat2))
    long1 = radians(float(long1))
    long2 = radians(float(long2))
    
    dlon = long2 - long1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    r = 6371
    return(c * r)
def shortest_dist(list_name):
    list_name.sort()
    return(list_name[0])
```

```python
counter_tract = 0
closest_library_km = []
while counter_tract < len(census_tract_lat):
    counter_library = 0
    library_dis = []
    while counter_library < len(library_lat):
        library_dis.append(determine_distance(census_tract_lat[counter_tract],library_lat[counter_library],census_tract_long[counter_tract],library_long[counter_library]))
        counter_library += 1
    counter_tract += 1
    closest_library_km.append(shortest_dist(library_dis))
print(closest_library_km)
    
```

### I then used the U.S. Census API with data from the 2018  5- year American Community Survey to determine the rate of highschool graduation by the age of 25, what percent of the population had broadband internet access at home, and the percent of people below the poverty line in each of the five remaining tracts.

```python
! pip install requests
```

```python
import requests
```

```python
#Highschool graduation rate by 25
base_url_hs_first = 'https://api.census.gov/data/2018/acs/acs5/profile?get=DP02_0066PE&for=tract:'
base_url_hs_sec = '&in=state:39%20county:153'
#&key=YOUR_KEY_GOES_HERE

# percent with internet connection
base_url_internet_first = 'https://api.census.gov/data/2018/acs/acs5/profile?get=DP02_0152PE&for=tract:'
base_url_internet_sec = '&in=state:39%20county:153'
#&key=YOUR_KEY_GOES_HERE

#percent below poverty level
base_url_poverty_first = 'https://api.census.gov/data/2018/acs/acs5/profile?get=DP03_0128PE&for=tract:'
base_url_poverty_sec = '&in=state:39%20county:153'
#&key=YOUR_KEY_GOES_HERE

```

```python
hs_graduation_rates = []
internet_connection = []
poverty_level = []

for tract in five_tracts:
    
    url_hs = base_url_hs_first + str(tract) + base_url_hs_sec
    url_internet = base_url_internet_first + str(tract) + base_url_internet_sec
    url_poverty = base_url_poverty_first + str(tract) + base_url_poverty_sec
    
    hs = requests.get(url_hs)
    internet = requests.get(url_internet)
    poverty = requests.get(url_poverty)
    
    hs_json = hs.json()
    internet_json = internet.json()
    poverty_json = poverty.json()
    
    hs_graduation_rates.append(hs_json[1][0])
    internet_connection.append(internet_json[1][0])
    poverty_level.append(poverty_json[1][0])
```
